## 2. Analysis: 

```python
# To-Do create a summary about each cluster : 
# percentages of each target (actual Category) within each cluster 
# + Percentages of each predicted class within the cluster 
# results should be something like : 
# For clusterk : {'true_class0_percetage': 0.0,
#                 'true_class1_percetage': 0.0,
#                  'true_class2_percetage': 0.0,
#                   'true_class3_percetage': 0.0,
#                   'true_class4_percetage': 0.0,
#                   'predicted_class0_percetage': 0.0,
#                   'predicted_class1_percetage': 0.0,
#                   'predicted_class2_percetage': 0.0,
#                   'predicted_class3_percetage': 0.0,
#                    'predicted_class4_percetage': 0.0}

true_classes = df['Category'].unique()
predicted_classes = df['Predicted_class'].unique()

# Initialize a dictionary to store the summary for each cluster
cluster_summary = {}

# Loop through each cluster and calculate percentages
for cluster in df['cluster'].unique():
    cluster_data = df[df['cluster'] == cluster]
    
    # Calculate the percentage of each true class within this cluster
    true_class_percentages = {f'true_class{cls}_percentage': (cluster_data['Category'] == cls).mean() * 100 for cls in true_classes}
    
    # Calculate the percentage of each predicted class within this cluster
    predicted_class_percentages = {f'predicted_class{cls}_percentage': (cluster_data['Predicted_class'] == cls).mean() * 100 for cls in predicted_classes}
    
    # Combine both true and predicted class percentages
    cluster_summary[cluster] = {**true_class_percentages, **predicted_class_percentages}

# Print the summary for each cluster
for cluster, summary in cluster_summary.items():
    print(f"Cluster {cluster}:")
    for key, value in summary.items():
        print(f"{key}: {value}%\n")
```

## 3. Explanations and insights :

As can be seen from the result, the spread of the predicted and true class percentages match - which indicates that the model has perfomed well and correctly predicted the class distribution.
